# PyPDF2 Extract Text from PDF

```python
# importing required modules
import PyPDF2
```

```python
import os

# assign directory
directory = "C:/Users/TechnologyDataServic/OneDrive - TCG3PL/Documents/Lori's Gifts/to_be_processed/pdfs"
iterations = 0
# iterate over files in
# that directory
for filename in os.scandir(directory):
    pdfFileObj = open(filename, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    pageObj = pdfReader.getPage(0)
    text_file = pageObj.extractText()
    with open('sep23.txt', 'a') as f:
        f.write(text_file)
        pdfFileObj.close()
        iterations +=1
print(f"We processed", iterations, "invoices for you.")
```
